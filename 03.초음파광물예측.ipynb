```python
from keras.models import Sequential, load_model
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.preprocessing import LabelEncoder
import pandas as pd
import numpy
import tensorflow as tf
# seed 값 설정
seed = 0
numpy.random.seed(seed)
tf.random.set_seed(3)
df = pd.read_csv('./data/sonar.csv', header=None)
```

```python
dataset = df.values
X = dataset[:,0:60].astype(float)
Y_obj = dataset[:,60]
e = LabelEncoder()
e.fit(Y_obj)
Y = e.transform(Y_obj)
```

```python
# 데이터를 배열 형태로 변환합니다.
#  - df.values: DataFrame을 numpy 배열로 변환하여 데이터셋을 생성합니다.
dataset = df.values

# 입력 데이터(X)와 출력 데이터(Y_obj)로 나눕니다.
#  - X: 데이터셋의 0번째부터 59번째 열까지 (입력 변수 60개)
#  - .astype(float): 데이터를 float형으로 변환하여 사용합니다.
X = dataset[:, 0:60].astype(float)

# Y_obj: 데이터셋의 60번째 열 (출력 변수)
Y_obj = dataset[:, 60]

# LabelEncoder를 사용하여 문자열 레이블을 숫자형으로 변환합니다.
e = LabelEncoder()  # LabelEncoder 객체를 생성합니다.
e.fit(Y_obj)        # Y_obj에 있는 문자열 레이블(출력 변수)을 학습시킵니다.

# Y_obj의 문자열 레이블을 숫자형 레이블로 변환합니다.
Y = e.transform(Y_obj)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=seed)
model = Sequential()
model.add(Dense(24,  input_dim=60, activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
model.compile(loss='mean_squared_error',
            optimizer='adam',
            metrics=['accuracy'])
```

```python
model.fit(X_train, Y_train, epochs=130, batch_size=5)
```

```python
import pickle  # 객체를 파일에 저장하거나 파일에서 불러올 때 사용하는 라이브러리

# 모델을 'my_model.pkl' 파일로 저장합니다.
#  - pickle.dump: 객체를 파일에 저장하는 함수
#  - model: 저장할 객체 (여기서는 학습된 모델)
#  - open('my_model.pkl', 'wb'): 'my_model.pkl' 파일을 쓰기 모드('wb', 바이너리 쓰기 모드)로 엽니다.
pickle.dump(model, open('my_model.pkl', 'wb'))
```

```python
# 'my_model.pkl' 파일에서 저장된 모델을 불러옵니다.
#  - pickle.load: 파일에서 객체를 불러오는 함수
#  - open('my_model.pkl', 'rb'): 'my_model.pkl' 파일을 읽기 모드('rb', 바이너리 읽기 모드)로 엽니다.
model = pickle.load(open('my_model.pkl', 'rb'))

# 불러온 모델을 사용하여 테스트 데이터를 평가합니다.
#  - model.evaluate: 불러온 모델로 X_test(입력 데이터)와 Y_test(레이블 데이터)를 평가합니다.
#  - [1]: 평가 결과에서 정확도(accuracy)를 가져옵니다.
print("\n Test Accuracy: %.4f" % (model.evaluate(X_test, Y_test)[1]))
```
